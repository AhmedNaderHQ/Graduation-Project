```python
from deepface import DeepFace
import pandas as pd
import cv2
```

Face Matching & Face Analyze

```python
# Load the known image
known_img = cv2.imread("images 1/Mazen Alasas.jpg")

# Check if the image loaded successfully
if known_img is None:
    print(f"Error: Unable to load image from {known_img}")
    exit()

# Use the webcam
cap = cv2.VideoCapture(0)

frame_num = 0
same_person = None

# Check the identity of the person in 30 frames
while frame_num < 30:
    ret, frame = cap.read()

    # if not ret:
    #     print("Error: Unable to read frame from camera.")
    #     break

    # cv2.imshow("Frame", frame)

    try:
        # Use img_path correctly
        same_person = DeepFace.verify(img1_path = known_img, img2_path = frame)["verified"]
        print(f'Frame number: {frame_num + 1} - Verified: {same_person}')
    except:
        pass

    if same_person:
        break

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_num += 1

cap.release()
# cv2.destroyAllWindows()

print("Same person verified:", same_person)

# Analyze the known image
# try:
#     analysis = DeepFace.analyze(img_path = frame)
#     print(analysis)
# except:
#     pass
```

Find Face & Face Analyze

```python
# Find a person in a dataset (if you have a set of images)

data_set_path ="images 1"  # Ensure this is the correct path to the folder

# Use the webcam again
cap = cv2.VideoCapture(0)

frame_num = 0
same_person = None


# Find a person in 30 frames
while frame_num < 30:
    ret, frame = cap.read()

    if not ret:
        print("Error: Unable to read frame from camera.")
        break

    cv2.imshow("Frame", frame)

    try:
    # Use img_path correctly
       
       same_person = DeepFace.find(db_path = data_set_path, img_path = frame, enforce_detection = False )
       print(f'Frame number: {frame_num + 1} - Find: {same_person}')
    except:
        pass   
    
    if same_person:
        break

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_num += 1

cap.release()
cv2.destroyAllWindows()


# # Analyze the known image
# try:
#     analysis = DeepFace.analyze(img_path=frame)
#     print(analysis)
# except:
#     pass
```
